```python
import matplotlib.pyplot as plt
from matplotlib.ticker import LinearLocator
from matplotlib.ticker import MaxNLocator
from matplotlib import cm
from mpl_toolkits.mplot3d import Axes3D

import numpy as np

from projections import project_points2D
```

```python
points = np.array([[1, 0, 0], [0, 0, 1], [0, 1, 0]])

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

surf = ax.plot_trisurf(points[:,0], points[:, 1], points[:, 2], cmap=cm.jet, color='r', alpha=0.6)

ax.scatter(points[0,0], points[0, 1], points[0, 2], color='r')
ax.scatter(points[1,0], points[1, 1], points[1, 2], color='b')
ax.scatter(points[2,0], points[2, 1], points[2, 2], color='g')


ax.xaxis.set_major_locator(MaxNLocator(5))
ax.yaxis.set_major_locator(MaxNLocator(6))
ax.zaxis.set_major_locator(MaxNLocator(5))

ax.set_ylabel("y")
ax.set_xlabel('x')
ax.set_zlabel('z')

fig.tight_layout()
plt.title('2-simplex in the 3 dimensional space')

plt.show()
```

```python
proj_points = project_points2D(points)
start_point = proj_points[0, :]

points = np.vstack([proj_points, start_point])
points
```

```python
fig, ax = plt.subplots()

plt.plot(points[:, 0], points[:, 1])
ax.scatter(points[0,0], points[0, 1], color='r')
ax.scatter(points[1,0], points[1, 1],  color='b')
ax.scatter(points[2,0], points[2, 1], color='g')
plt.fill(points[:, 0], points[:, 1], color='b', alpha=0.5)

ax.set_ylabel('y')
ax.set_xlabel('x')

plt.title('Projected on the 2-dim plane')
```
