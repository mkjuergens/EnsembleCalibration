```python
import torchvision
import torch
from torch.utils.data import DataLoader
import matplotlib.pyplot as plt
import numpy as np
```

```python
dataset = torchvision.datasets.MNIST(root='data/', download=True, transform=torchvision.transforms.ToTensor())
```

```python
dataloader = DataLoader(dataset, batch_size=32, shuffle=True)
```

```python
def imshow(img):
    img = img / 2 + 0.5  # unnormalize
    npimg = img.numpy()
    fig, ax = plt.subplots(figsize=(12, 8))
    plt.imshow(np.transpose(npimg, (1, 2, 0)))
    plt.show()
```

```python
dataiter = iter(dataloader)
images, labels = next(iter(dataloader))

plt.figure(figsize=(7, 7))
# Show images
noise = torch.randn(images[0].shape) * 0.0
noisy_image = images[0] + noise
plt.imshow(noisy_image.numpy().squeeze(), cmap='gray')

# Add noise to the first image
noise = torch.randn(images[0].shape) * 0.8  # Adjust noise level as needed
noisy_image = images[0] + noise
noisy_image = torch.clamp(noisy_image, 0.0, 1.0)  # Ensure pixel values are valid

# Show the noisy image
plt.figure(figsize=(7,7))
plt.imshow(noisy_image.numpy().squeeze(), cmap='gray')
plt.show()
```
