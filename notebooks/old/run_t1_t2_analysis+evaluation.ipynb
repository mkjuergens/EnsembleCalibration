# Test Simulations for the Binary Cone Experiment

## 1. Test $H_0$ hypothesis

```python
# alpha values to run the test with
alphas = [0.05, 0.13, 0.21, 0.30, 0.38, 0.46, 0.54, 0.62, 0.70, 0.78, 0.87, 0.95]
# number of resamplings
n_resamples = 50
n_samples = 10000
```

The first test examines the case where the null hypothesis is true, i.e.  there exists a calibrated convex combination within the cone spanned by the two ensmeble members.


```python
fig, ax = plt.subplots(figsize=(7, 4))
# visualize convex combination for different degrees of weight functions
degs = [0, 1, 2, 3, 4, 5]
for deg in degs:
    x_inst, p_probs, y_labels, p_bar = cone_experiment_h0(n_samples=1000, deg_fct=deg)

    if deg == 0:
        ax.scatter(x_inst.squeeze(), p_probs[:, 0, 0], alpha=0.2, s=1, color="black")
        ax.scatter(x_inst.squeeze(), p_probs[:, 1, 0], alpha=0.2, s=1, color="black")
    
    ax.scatter(x_inst.squeeze(), p_bar[:, 0], s=1, label=f'$\text{deg}_w={{{deg}}}$')
plt.title("Analysis of the null hypothesis")
ax.set_xlabel("x")
ax.set_ylabel("$(\hat{p}(x))_1$")


```

```python
results = _simulation_h0_cone(config_new_mlp_binary, 
                              n_samples,
                                n_resamples,
                                alphas)
```

```python
results
```

```python
# plot the results
fig, ax = plt.subplots( 1, len(results), figsize=(9, 4))

for i, key in enumerate(results):
        ax[i].plot(alphas, results[key])
        ax[i].set_title(f"{key}")
        ax[i].set_xlabel(r"$\alpha$")
        ax[i].set_ylabel(r"Type $1$ error")
        ax[i].plot(alphas, alphas, "--")
plt.tight_layout()
```

# 2. Test $H_a$ hypothesis

```python
fig, ax = plt.subplots(figsize=(7, 4))
# visualize convex combination for different degrees of weight functions
fracs_in = [0, 0.25, 0.5, 0.75, 1]
for i, frac in enumerate(fracs_in):
    x_inst, p_probs, y_labels, p_bar = cone_experiment_h1(n_samples=1000, frac_in=frac)

    if frac == 0:
        ax.scatter(x_inst.squeeze(), p_probs[:, 0, 0], alpha=0.2, s=1, color="black")
        ax.scatter(x_inst.squeeze(), p_probs[:, 1, 0], alpha=0.2, s=1, color="black")
    
    ax.scatter(x_inst.squeeze(), p_bar[:, 0], s=1, label=f'$\overline{{p}}_{{{i}}}(x))_1$')
    ax.vlines(frac, ymin=0, ymax=0.5*frac, color="red", linestyle="--")

plt.title("Analysis of the alternative hypothesis \n in dependence of the fraction within the cone")
ax.set_xlabel("x")
ax.set_ylabel("(\hat{p}(x))_1")
```

### Case 1: Predictions lie completely outside convex hull

```python
frac_in = 0.0

results = _simulation_h1_cone(
    config_new_mlp_binary,
    n_samples=n_samples,
    n_resamples=n_resamples,
    alpha=alphas,
    experiment=cone_experiment_h1,
    frac_in=frac_in
)
```
