```python
import matplotlib.pyplot as plt
import mpltern
```

```python

fig = plt.figure(figsize=(10.8, 4.8))
fig.subplots_adjust(left=0.075, right=0.85, wspace=0.3)

ax = fig.add_subplot(1, 2, 1, projection="ternary")


ax.set_tlabel('$1$')
ax.set_llabel('$2$')
ax.set_rlabel('$3$')

# scatter
ax.scatter([0.4], [0.3], [0.7], c='C0', s=100)
ax.scatter([0.6], [0.6], [0.8], c='C1', s=100)

ax.axtline(0.2, c='C0')
ax.axlline(0.3, c='C1')
ax.axrline(0.4, c='C2')

ax.axtspan(0.3, 0.5, fc='C0', alpha=0.2)
ax.axlspan(0.4, 0.6, fc='C1', alpha=0.2)
ax.axrspan(0.5, 0.7, fc='C2', alpha=0.2)
```
